# Undersampling

```python
normal = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=473)  #it will select 473 values RANDOMLY from the normal data
normal_sample.shape
```

```python
#concating our normal samples to our fraudulent samples

new_data = pd.concat([normal_sample,fraud],ignore_index=True) # while concatinating we don't want it to genereate random indexes, we want to keep the original ones
new_data['Class'].value_counts()
```

```python
sns.countplot(new_data['Class'])
```

```python
#copying our newly created dataset into X and y

X = new_data.drop('Class',axis=1)
y = new_data['Class']
```

```python
#perform training and testing on the new x and y

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
# After undersampling is done, Now the dataset is ready to apply our models
```

```python
y_pred1 = log.predict(X_test)
```
